1: In simple words, what are React Hooks? Create a small component that shows a button. When the button is clicked, toggle the text between **"On"** and **"Off"**.

**React Hooks** are special functions introduced in React that allow developers to use state and other React features inside functional components without writing class components. Hooks make React code simpler, cleaner, and easier to reuse.

One of the most commonly used Hooks is **`useState`**, which is used to create and manage state in a functional component.

### Features of React Hooks

- Allow state management in functional components.
- Make code simpler and easier to understand.
- Improve code reusability.
- Reduce the need for class components.
- Support side effects using `useEffect`.

### React Code

```jsx
import React, { useState } from "react";

function ToggleButton() {
    const [status, setStatus] = useState("Off");

    const toggleStatus = () => {
        setStatus(status === "Off" ? "On" : "Off");
    };

    return (
        <div>
            <h2>{status}</h2>
            <button onClick={toggleStatus}>
                Toggle
            </button>
        </div>
    );
}

export default ToggleButton;
```

2: Make a counter with a button. Whenever the count changes, log **"Count changed"** using `useEffect`.

The `useState` Hook is used to store and update the counter value, while the `useEffect` Hook is used to perform side effects. In this example, whenever the `count` value changes, `useEffect` runs and logs **"Count changed"** to the browser console.

### React Code

```jsx
import React, { useState, useEffect } from "react";

function Counter() {
    const [count, setCount] = useState(0);

    useEffect(() => {
        console.log("Count changed");
    }, [count]);

    return (
        <div>
            <h2>Count: {count}</h2>

            <button onClick={() => setCount(count + 1)}>
                Increment
            </button>
        </div>
    );
}

export default Counter;
```

3: Write a component that fetches users from `https://jsonplaceholder.typicode.com/users` only once on mount and shows how many users were loaded.

The `useEffect` Hook is used to perform side effects such as fetching data from an API. By providing an empty dependency array (`[]`), the effect runs only once when the component is mounted. The fetched users are stored using the `useState` Hook, and the total number of users loaded is displayed.

### React Code

```jsx
import React, { useState, useEffect } from "react";

function UserCount() {
    const [users, setUsers] = useState([]);

    useEffect(() => {
        fetch("https://jsonplaceholder.typicode.com/users")
            .then((response) => response.json())
            .then((data) => {
                setUsers(data);
            })
            .catch((error) => {
                console.error("Error fetching users:", error);
            });
    }, []);

    return (
        <div>
            <h2>Users Loaded: {users.length}</h2>
        </div>
    );
}

export default UserCount;
```

5: Build a small form with a text input. If the input is empty, show **"Please type your name"**. If it has text, show **"Hello, <name>"**.


```jsx
import React, { useState } from "react";

function NameForm() {
    const [name, setName] = useState("");

    return (
        <div>
            <h2>Name Form</h2>

            <input
                type="text"
                placeholder="Enter your name"
                value={name}
                onChange={(e) => setName(e.target.value)}
            />

            <p>
                {name.trim() === ""
                    ? "Please type your name"
                    : `Hello, ${name}`}
            </p>
        </div>
    );
}

export default NameForm;
```


6: You have a theme value ("light" or "dark") that is needed by a deep child button. Show how to pass it without sending `theme` through every parent prop.


In React, **Context API** is used to share data between components without passing props through every intermediate component. This avoids **prop drilling**, where props have to be passed through multiple levels of components even if they do not use the data.

In this example, the `theme` value is stored in a context and accessed directly by the deep child button using the `useContext` Hook.

### Step 1: Create the Context

```jsx
import React from "react";

const ThemeContext = React.createContext();

export default ThemeContext;
```

### Step 2: App Component

```jsx
import React from "react";
import ThemeContext from "./ThemeContext";
import Parent from "./Parent";

function App() {
    const theme = "light";

    return (
        <ThemeContext.Provider value={theme}>
            <Parent />
        </ThemeContext.Provider>
    );
}

export default App;
```

### Step 3: Parent Component

```jsx
import React from "react";
import Child from "./Child";

function Parent() {
    return <Child />;
}

export default Parent;
```

### Step 4: Child Component

```jsx
import React from "react";
import DeepChild from "./DeepChild";

function Child() {
    return <DeepChild />;
}

export default Child;
```

### Step 5: DeepChild Component

```jsx
import React, { useContext } from "react";
import ThemeContext from "./ThemeContext";

function DeepChild() {
    const theme = useContext(ThemeContext);

    return (
        <button>
            Theme: {theme}
        </button>
    );
}

export default DeepChild;
```

### Output

```text
Theme: light
```

 7: Make a counter with two buttons: **"+1"** and **"-1"**. Use `useReducer` instead of `useState`.

The `useReducer` Hook is used to manage state when state updates are more complex. It works by using a **reducer function** that receives the current state and an action, then returns the updated state. In this example, the counter can be increased or decreased using two buttons.

### React Code

```jsx
import React, { useReducer } from "react";

// Reducer function
function reducer(state, action) {
    switch (action.type) {
        case "increment":
            return { count: state.count + 1 };

        case "decrement":
            return { count: state.count - 1 };

        default:
            return state;
    }
}

function Counter() {
    const [state, dispatch] = useReducer(reducer, { count: 0 });

    return (
        <div>
            <h2>Count: {state.count}</h2>

            <button onClick={() => dispatch({ type: "increment" })}>
                +1
            </button>

            <button onClick={() => dispatch({ type: "decrement" })}>
                -1
            </button>
        </div>
    );
}

export default Counter;
```

### Output

```text
Count: 0

[+1] [-1]

After clicking +1:
Count: 1

After clicking -1:
Count: 0
```


8: Given an array `nums = [1, 2, 3, 4]` and a filter value `min`, show how to compute a filtered list only when `nums` or `min` change. Also pass a stable click handler to a child list item.


In React, the **`useMemo`** Hook is used to memoize computed values so that they are recalculated only when their dependencies change. The **`useCallback`** Hook is used to memoize functions, ensuring that the same function instance is reused between renders unless its dependencies change.

In this example:

- `useMemo` computes the filtered list only when `nums` or `min` changes.
- `useCallback` creates a stable click handler that is passed to the child component.

### Child Component

```jsx
import React from "react";

function ListItem({ number, onClick }) {
    return (
        <li onClick={() => onClick(number)}>
            {number}
        </li>
    );
}

export default ListItem;
```

### Parent Component

```jsx
import React, { useMemo, useCallback } from "react";
import ListItem from "./ListItem";

function NumberList() {
    const nums = [1, 2, 3, 4];
    const min = 3;

    // Compute filtered list only when nums or min changes
    const filteredNumbers = useMemo(() => {
        return nums.filter((num) => num >= min);
    }, [nums, min]);

    // Stable click handler
    const handleClick = useCallback((number) => {
        console.log("Clicked:", number);
    }, []);

    return (
        <div>
            <h2>Filtered Numbers</h2>

            <ul>
                {filteredNumbers.map((number) => (
                    <ListItem
                        key={number}
                        number={number}
                        onClick={handleClick}
                    />
                ))}
            </ul>
        </div>
    );
}

export default NumberList;
```

9: Create a controlled input for email. As the user types, show **"Email: <value>"**. Why do we keep the input's value in state?


A **controlled component** in React is an input element whose value is controlled by the component's state. The `useState` Hook stores the input value, and every time the user types, the state is updated. This keeps the displayed value and the input field synchronized.

### React Code

```jsx
import React, { useState } from "react";

function EmailInput() {
    const [email, setEmail] = useState("");

    return (
        <div>
            <h2>Email Form</h2>

            <input
                type="email"
                placeholder="Enter your email"
                value={email}
                onChange={(e) => setEmail(e.target.value)}
            />

            <p>Email: {email}</p>
        </div>
    );
}

export default EmailInput;
```

In React, the **`useRef`** Hook is used to create a reference to a DOM element. It allows direct access to the element without causing a re-render. In this example, clicking the button focuses the input field. A **React Fragment (`<>...</>`)** is used to return multiple adjacent elements without adding an extra wrapper element to the DOM.

### React Code

```jsx
import React, { useRef } from "react";

function FocusInput() {
    const inputRef = useRef(null);

    const handleFocus = () => {
        inputRef.current.focus();
    };

    return (
        <>
            <input
                type="text"
                ref={inputRef}
                placeholder="Enter text"
            />

            <button onClick={handleFocus}>
                Focus Input
            </button>
        </>
    );
}

export default FocusInput;
```

### Output

```text
[Enter text]

[Focus Input]

(When the button is clicked)

The cursor automatically moves to the input field.
```
